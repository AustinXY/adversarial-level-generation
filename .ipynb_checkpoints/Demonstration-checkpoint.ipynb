```python
import gym
from stable_baselines3 import PPO
from envs import ALGEnv, ALGDemoWrapper
from stable_baselines3.common.vec_env import DummyVecEnv, SubprocVecEnv
from stable_baselines3.common.utils import set_random_seed
import os
from IPython.display import Image, display
import numpy as np
```

```python
num_boxes = 1
alg_version = 0
dim_room = (7, 7)
train_mode = 'mlp'
alg_path = 'demo_checkpoints/alg_v1b_v0'
agent_lb_path = None
agent_ub_path = None
```

```python
alg_env = ALGEnv(dim_room=dim_room, num_boxes=num_boxes, train_mode=train_mode,
                 alg_version=alg_version, agent_lb_path=agent_lb_path, agent_ub_path=agent_ub_path)
alg_demo = ALGDemoWrapper(alg_env, alg_path=alg_path)
```

```python
init_map = np.zeros(dim_room)
box_pos = ()
box_target_pos = ()
player_pos = (5,5)

if box_pos != ():
    init_map[box_pos] = 2
if box_target_pos != ():
    init_map[box_target_pos] = 4
if player_pos != ():
    init_map[player_pos] = 5
print(init_map)
```

```python
if alg_demo.generate_episode_gif(init_map):
    display(Image(filename='temp/alg_episode.gif', width=200, height=200))
```
